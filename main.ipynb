```python
#Importing libraries
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

#Importing datasets
df_no_of_tourists=pd.read_csv("API_ST.INT.ARVL_DS2_en_csv_v2_10226141.csv",header=[2])
df_earning=pd.read_csv("API_ST.INT.RCPT.CD_DS2_en_csv_v2_10225013.csv",header=[2])
df_regional_stats=pd.read_csv("international-tourist-arrivals-by-world-region.csv",header=[0])


#Cleaning data
columns_to_drop = ['Indicator Name','Indicator Code','1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973','1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984','1985','1986','1987','1988','1989','1990','1991','1992','1993','1994','2017']
df_no_of_tourists.drop(columns_to_drop, inplace=True, axis=1)

df_earning.drop(columns_to_drop, inplace=True, axis=1)

#remove null values
df_no_of_tourists = df_no_of_tourists.replace(np.nan, 0)
df_earning=df_earning.replace(np.nan, 0)

#delete rows that are not a country
rows_to_be_deleted = [5,34,59,60,61,62,63,66,71,72,93,96,100,101,102,103,108,126,132,133,134,137,138,140,151,154,159,168,179,181,189,195,196,202,213,215,216,228,229,234,236,238,239,
247,257]
df_no_of_tourists.drop(df_no_of_tourists.index[rows_to_be_deleted],inplace=True)
df_earning.drop(df_earning.index[rows_to_be_deleted],inplace=True)

#Select all columns with valid data
columns_to_plot = ['1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']

df_no_of_tourists = df_no_of_tourists.apply(pd.to_numeric, errors='ignore')
df_earning= df_earning.apply(pd.to_numeric, errors='ignore')

#Calculate the total arrivals per year
tourists_per_year = [df_no_of_tourists['1995'].sum(),df_no_of_tourists['1996'].sum(),df_no_of_tourists['1997'].sum(),df_no_of_tourists['1998'].sum(),df_no_of_tourists['1999'].sum(),df_no_of_tourists['2000'].sum(),df_no_of_tourists['2001'].sum(),df_no_of_tourists['2002'].sum(),df_no_of_tourists['2003'].sum(),df_no_of_tourists['2004'].sum(),df_no_of_tourists['2005'].sum(),df_no_of_tourists['2006'].sum(),df_no_of_tourists['2007'].sum(),df_no_of_tourists['2008'].sum(),df_no_of_tourists['2009'].sum(),df_no_of_tourists['2010'].sum(),df_no_of_tourists['2011'].sum(),df_no_of_tourists['2012'].sum(),df_no_of_tourists['2013'].sum(),df_no_of_tourists['2014'].sum(),df_no_of_tourists['2015'].sum(),df_no_of_tourists['2016'].sum()]

#Calculate total revenue generated through tourism per year
revenue_per_year = [df_earning['1995'].sum(),df_earning['1996'].sum(),df_earning['1997'].sum(),df_earning['1998'].sum(),df_earning['1999'].sum(),df_earning['2000'].sum(),df_earning['2001'].sum(),df_earning['2002'].sum(),df_earning['2003'].sum(),df_earning['2004'].sum(),df_earning['2005'].sum(),df_earning['2006'].sum(),df_earning['2007'].sum(),df_earning['2008'].sum(),df_earning['2009'].sum(),df_earning['2010'].sum(),df_earning['2011'].sum(),df_earning['2012'].sum(),df_earning['2013'].sum(),df_earning['2014'].sum(),df_earning['2015'].sum(),df_earning['2016'].sum()]

tourists_per_year = [round(x) for x in tourists_per_year]
```

```python
#Plotting a multiple axis line chart to show evolution of tourism over the years
label_text='<b>International Tourism Statistics.'+ '<br>' + 'Number of arrivals and the revenue earned through tourism from 1995 to 2016 </b>'+ '<br>' + 'Source - https://data.worldbank.org/'
trace1 = go.Scatter(
     x=columns_to_plot,
    y = list(np.around(np.array(tourists_per_year),2)),
    name='Number of tourists per year'
)
trace2 = go.Scatter(
    x=columns_to_plot,
    y=revenue_per_year,
    name='Revenue earned per year',
    yaxis='y2'
)
data = [trace1, trace2]
layout = go.Layout(
    title=label_text,
    yaxis=dict(
        title='Number of arrivals (Billions)'
    ),
    yaxis2=dict(
        title='Tourism Revenue (Trillion USD)',
        
        
        overlaying='y',
        side='right'
    )
)
fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='multiple-axes-double.html')
```

```python
#Preparing data for graph
df_test=df_regional_stats
df_test.drop('Code', inplace=True, axis=1)
df_africa=df_test[df_test['Entity'].str.contains('Africa')]
df_africa.reset_index(drop=True)
df_americas=df_test[df_test['Entity'].str.contains('Americas')]
df_americas.reset_index(drop=True)
df_asia_pacific=df_test[df_test['Entity'].str.contains('Asia & Pacific')]
df_asia_pacific.reset_index(drop=True)
df_europe=df_test[df_test['Entity'].str.contains('Europe')]
df_europe.reset_index(drop=True)
df_middle_east=df_test[df_test['Entity'].str.contains('Middle East')]
df_middle_east.reset_index(drop=True)


africa_list=df_africa[' (arrivals)'].where(df_africa['Entity'] == 'Africa').tolist()
americas_list=df_americas[' (arrivals)'].where(df_americas['Entity'] == 'Americas').tolist()
asia_pacific_list=df_asia_pacific[' (arrivals)'].where(df_asia_pacific['Entity'] == 'Asia & Pacific').tolist()
europe_list=df_europe[' (arrivals)'].where(df_europe['Entity'] == 'Europe').tolist()
middle_east_list=df_middle_east[' (arrivals)'].where(df_middle_east['Entity'] == 'Middle East').tolist()
year_list=df_test['Year'].where(df_test['Entity'] == 'Africa').tolist()
```

```python
#Plotting scattered area chart to show the distribution of percentage of total number of arrivals by world region.
Africa = dict(
    x=year_list,
    y=africa_list,
    mode='lines',
    line=dict(width=0.5,
              color='rgb(204, 204, 255)'),
    stackgroup='one',
    groupnorm='percent',
    name='Africa'
)
America = dict(
    x=year_list,
    y=americas_list,
    mode='lines',
    line=dict(width=0.5,
              color='rgb(240, 31, 31)'),
    stackgroup='one',
    name='America'
)
AsiaPacific = dict(
    x=year_list,
    y=asia_pacific_list,
    mode='lines',
    line=dict(width=0.5,
              color='rgb(240, 240, 31)'),
    stackgroup='one',
    name='Asia and Pacific'
)
Europe = dict(
    x=year_list,
    y=europe_list,
    mode='lines',
    line=dict(width=0.5,
              color='rgb(33, 75, 174)'),
    stackgroup='one',
    name='Europe'
)
MiddleEast = dict(
    x=year_list,
    y=middle_east_list,
    mode='lines',
    line=dict(width=0.5,
              color='rgb(79,154,97)'),
    stackgroup='one',
    name='Middle East'
)

data = [Africa, America, AsiaPacific, Europe,MiddleEast]
layout = go.Layout(
    title='<b>International Tourist Arrivals By World Region</b>'+ '<br>Source : UNWTO Tourism Barometer, Volume 15 Advance Release January 2015',
    showlegend=True,
    xaxis=dict(
        type='linear',
        range=[1950, 2016],
        dtick=10,
    ),
    yaxis=dict(
        type='linear',
        range=[1, 100],
        dtick=20,
        ticksuffix='%'
    )
)
fig = dict(data=data, layout=layout)
py.iplot(fig, filename='stacked-area-plot-norm.html')
```
